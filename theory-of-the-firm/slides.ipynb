## Tранзакционни разходи (Coase)

Фирмите извършват действия в рамките на фирмата, защото извършването им на пазара може да бъде скъпо:

- договори с доставчици;
- съдебни дела, когато нещата се объркат;
- търсенето на служители и много други задачи не са безплатни. 

Когато е твърде скъпо да се извършват тези дейности отново и отново, ние можем да формираме по-стабилен набор от отношения между средствата за производство и продукцията, където йерархичното ръководство замества пазарните транзакции.

## Извод

Не можем да обясним фирмата само с различните стимули на работници и мениджъри, производители и доставчици и т.н. Ако можеха свободно да се договарят, пазарните участници биха предприели всяко действие, което максимизира индивидуалната им полза и след това да разделят ползите по взаимноизгоден начин. Теоремата на Коус ни казва, че фирмите същестуват за да премахват транзакционите разходи в преговорите и така да създават възможността за взаимноизгодни споразумения.


## Tранзакционни разходи (Williamson)

<br>

За да съществуват фирмите, към тях трябва да са налични усвояеми квази-ренти, които се появяват в резултат на специфична обща дейност във фирмата (*appropriable quasi-rents*). 


## Tранзакционни разходи (Williamson)

<br>


AQRs възникват тогава, когато общите действия имат по-високи нетни приходи, отколкото разходи и дават възможност да се преговаря за тяхното разделяне.

Когато две фирми създават AQRs от обща специфична дейност, казваме, че е настъпила **фундаментална трансформация**.


## Въпрос

<br>

::: {.incremental}

- Може ли да има фундаментална трансформация при съвършена конкуренция?

- Не: ако един доставчик не желае да участва в общата дейност, просто може да бъде намерен следващ, за същата цена и печалбата за него ще бъде равна на $0$ заради конкуренцията.

- При съвършена конкуренция няма нужда да мислим за **специални отношения**.

:::

## Tранзакционни разходи (Williamson)

<br>

- AQRs са важни за съществуването на фирмата ако по някакви причини фирмите не могат да преговарят за всяка възможност. 

- Не може да бъдат специфицирани всички подробности по обастоятелствата, които могат да възникнат в бъдеще по един договор: договорите са несъвършени.


## Tранзакционни разходи (Williamson)

<br>

Причините договорите да са несъвършени могат да се сведат до **ограничена рационалност**: никой не може да предвиди всички бъдещи обстоятелства.

<br>

Тогава страните на пазара трябва да предоговарят, докато във фирмата мениджърът просто ще вземе решение.

## Tранзакционни разходи (Williamson)

<br>

Извод: дейностите вътре във фирмата са тези, които изискват честа адаптация към външни промени. Затова някои фирмени дейности дейности могат да изглеждат сложни в сравнение с пазарните транзакции: *причината* тези дейности да се извършват във фирмата е, че те изискват мениджърски контрол за адаптация към промени.

Ако тези дейности бяха изнесени извън фирмата, това щеше да доведе до чести преговори, пазарене и предговаряне. 


## Tранзакционни разходи (Williamson)

<br>

Извод: възникването на непредвидени обстоятелства не гарантира, че мениджърите ще опитват да налагат волята си. По-скоро всеки заинтересуван ще опитва да им влияе спрямо собствените си интереси, което ще доведе до загуба на ресурси. 

Високите *разходи за влияние* намаляват ползата от фирмената интеграция.


## Проблемът за задържането (hold-up)

*Suppose that an electricity generator has strong cost-based incentives to locate near a coal mine. Building a new generator involves a sunk investment. However, once this investment is sunk, the generator firm will find itself in a bilateral monopoly situation vis- à-vis the coal mine. The electricity generator can sign a contract with the coal mine before investing in a generator. However, after the generator is built, it becomes a sunk cost. The coal mine will have an incentive to seek some reinterpretation or renegotiation of the contract that would allow it to receive a higher price for coal. If the electricity generator anticipates this “hold-up” situation, it may simply decide not to make the investment.* (Aghion and Holden, 2011)

## Права на собственост

*It is common to see the firm characterized by the power to settle issues by fiat, by authority, or by disciplinary action superior to that available in the conventional market. This is delusion. The firm does not own all its inputs. It has no power of fiat, no au- thority, no disciplinary action any different in the slightest degree from ordinary market contracting between any two people. I can “punish” you only by withholding future business or by seeking redress in the courts for any failure to honor our exchange agree- ment. This is exactly all that any employer can do. He can fire or sue, just as I can fire my grocer by stopping purchases from him or sue him for delivering faulty products.* Alchian and Demsetz, (1972).

## Права на собственост

Каква е разликата?

::: {.incremental}

- Когато "уволня" моят доставчик, той запазва своята собственост (машини, ресурси, служители, капитал).

- Когато уволня моят служител, моята фирма ги запазва.

- Това е основното в теорията за правата на собственост, или "residual control rights". 

:::

## Права на собственост

<br>


::: {.incremental}

- Основният въпрос тук не е "да правя или да купувам?", а "какво е фирмата?"

- Отговорът: каквото и да прави формата в даден период, в началото тя протежава определени активи и може да ги използва както иска.

:::

## Права на собственост

Теоремата на Коус ни казва, че правата на собственост нямат значение, но това не е вярно при две условия:

::: {.incremental}

- Наличие на непълни/несъвършени договори: тогава има значение кой притежава активите.

- Наличие на специфични инвестиции, които променят стойността на фирмените активи (проблемът за задържането): активите трябва да са притежавани от този, чиято инвестиция е най-чувствителна към промени в собствеността.

:::

## Ресурсна перспектива

В теориите за транзакционните разходи и правата на собственост не става дума за такива неща като "фирмената култура".

## Фирмата от гледна точка на знанието

Знанието има два компонента: практическо знание (know-how) и информация.

::: {.incremental}

- Практическото знание включва "акумулираните практически умения или експертиза , която позволява на някого да върши работа гладко и ефикасно" (Eric Von Hippel, The Sources of Innovation).

- Информацията може лесно да бъде комуникирана, когато фирмата има правила за кодиране и разкодиране.

## Фирмата от гледна точка на знанието

<br>

Фирмите позволяват на индивидите в тях да развиват и комбинират тези два типа знание в полезни и трудни за репликиране системи. 

Тъй като тези системи не са собственост на един конкретен индивид, те са поддържани, развивани и използвани в границите на фирмата, като й позволяват да създава иновации и да развива предимство пред конкурентите.


## Фирмата от гледна точка на знанието

Знанието позволява на формата да решава проблеми:

::: {.incremental}

- прости проблеми се решават по принципа за минимирането на разходите: "какъв материал да избера?"

- за средно трудни проблеми, решението може да бъде намерено, използвайки фирмената йерархия;

-  за най-трудни проблеми, може да се използват правила за търсене на решение, за спиране на търсенето, предварително изработени процедури или създаване на нови.

:::


## Фирмата от гледна точка на знанието

::: {.incremental}

- Акумулирането на знание става благодарение на постоянството на отношенията във фирмите, в сравнение с пазарните транзакции.

- Оказва се, че възможността да се решават проблеми с помощта на специфично знание е един от основните ресурси, които фирмите не могат винаги да купуват на пазарите.

:::

## Фирмата от гледна точка на знанието

::: {.incremental}

- Интегрираните производители на чипове имат повече пазарен успех при продукти, които изиксват комплексно знание за решаване на сложни проблеми Macher (Management Sciences, 2006);

- Учените във фармацевтичните компании имат повече патенти, когато работят в голяма диверсифицирана фирма, вместо в по-малка, специализирана (Henderson and Cockburn (PNAS 1996)) 

:::

## Фирмата от гледна точка на знанието

Знанието е скъпо: 

::: {.incremental}

- фирмите трудно предвиждат какви проблеми ще трябва да решават в бъдеще и оттам - какви приоритети, свързани със знанието да преследват;
- те могат да се обвържат с решения, зависими от миналото и това да ги обрече на изпълнение на рутинни дейности (path dependency);
- целейки да прикрият знанието, което имат, може да се откажат от потенциално полезни партньорства.
:::

## Фирмата от гледна точка на проблема за принципала и агента

Служителите трябва да бъдат наблюдавани и съществува информационна асиметрия.

Holmstrom и Milgrom: когато служителите вършат повече от една дейност (multitasking) е трудно да бъдат стимулирани оптимално; ако усилията на един доставчик не може да бъдат директно наблюдавани, може да не съществува договор с няколко различни цели, който да бъде ефикасен.


## Критиката на Maskin и Tirole

::: {.incremental}

- Дотук много често казвахме, че фирмите същестуват заради невъзможността от изработването на съвършени договори.

- Maskin и Tirole показват, че фирмите, без да знаят бъдещите обстоятелства, могат да сключват договори, **като че ли ги знаят**.

:::

## Критиката на Maskin и Tirole

<br>

... Такъв договор вместо "ако се случи $X$, ще правим $Y$", би имал следната форма: "ако нещо неочаквано се случи и максималната полза от междуфирмените отношения в бъдеще се промени на $Z$, ще организиране такива компромиси, че да можем да си разделим пълния размер на $Z$.
