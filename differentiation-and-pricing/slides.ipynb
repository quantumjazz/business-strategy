## Продуктова диференциация

Demsetz (1973, 1974): традиционно има два начина да се гледа на това как конкуренцията води до увеличаване на печалбата:

- ограничаване на съперничеството за намаляване на ценовата конкуренция;
- възползване от конкурентно предимство за създаване на повече 'икономическа стойност', където под 'икономическа стойност' се разбира разликата между желанието на потребителите да платят за едно благо и фирмените разходи.

## Още един дълъг цитат:

*If we consider a class of goods as being
typified by a set of (desirable) characteristics,
then two varieties are vertically differentiated
when the first contains more of some or
all characteristics than the second, so that
all rational consumers given a free choice
would opt for the first. They are horizontally
differentiated when one contains more of
some but fewer of other characteristics,
so that two consumers exhibiting different
tastes offered a free choice would not
unambiguously plump for the same one.*
(Waterson, 1989: 2).

Ако моделираме хоризонтална диференциация:

* две фирми се конкурират на пазар, където потребителите се различават в оценката си за идеалните качества на един продукт (примери: минерална вода, автомобили, ресторанти)

## Хоризонтална диференциация

* два ефекта: 
    + фирмата се разграничава от своя конкурент, което намалява ценовото съперничество;
    + фирмата се доближава до аст от потребителите, които намират продукта за по-привлекателен, но я отдалечава от останалите, които го приемат за по-малко желателен.


## Хоризонтална диференциация

**Комбинираният ефект е нетно конкурентно изоставане**

## Вертикална диференциация

Ако моделираме вертикална диференциация:

* тогава фирмата търси движение към по-добра позиция от гледна точка на *качеството*, което се отнася за *всички* потребители на пазара

## Вертикална диференциация

* това не означава непременно, че фирмата има предимство в създаването на 'икономическа стойност', защото
    + потребителите се различават в склонноста им да платят за качеството;
    + по-високото качество сигурно ще изисква по-високо ниво на фирмените разходи.

## Вертикална диференциация

Но понеже нашата дефиниция за вертикална диефенциация изисква увеличаване на качеството на някои аспекти на продукта без влошаването на други, то нарастването на качеството трябва да води до нарастване на потребителското желание да платят и на фирмените разходи.

Затова, за фирмите, които са по-близо до оптималното съотношение между цена и качество **комбинираният ефект е нетно конкурентно предимство**

## Един извод

Достатъчен е малък брой доставчици за да може потребителите да извлекат ползи от конкуренцията. Ако има доминиращ доставчик, който предлага повече стойност, разумните купувачи го избират за повече, но не всичките си нужди, за да държат останалите доставчици наоколо. Това е достатъчно за наличието на конкурентен натиск върху основния доставчик.


## Повторения на ценовата дилема

Възможни стратегии при многобройни итерации на играта:

1. Винаги избирай по-ниската цена.
2. Избери по-високата цена в първия рунд на играта. Всеки следващ рунд, в който опонентът избере високата цена, избери високата цена. Ако опонентът избере ниската цена веднъж, избирай ниската цена до края на играта. Това е известно като **the grim trigger**.


## Повторения на ценовата дилема - продължение

3. Избери високата цена в първия рунд на играта. Всеки следващ рунд избирай това, което опонентът ти е избрал в предишния.  Това е известно като **tit for tat**.

## Забранени споразумения и агресивно ценообразуване (predatory pricing)

Тези споразумения заменят конкуренцията с пазарене в затворени хотелски стаи. Участниците в тези преговори се договарят как да разделят печалбите от споразумението помежду си. В дългосрочен план това означава наблюдение на цените и количествата между тях.


## Забранени споразумения и агресивно ценообразуване (predatory pricing)

Ако фирмите имат ограничен капитал, намаляването на цената може да изведе от пазара тези, които имат по-високи разходи. При две фирми, останалият се превръща в монополист. Ако има достатъчно време, той може да възстанови разходите, направени за агресивнно ниските цени.

**The winner's curse.**
