## Основни идеи
- Силата на имитацията
- Ефективност на операциите
- Конкуретна конвергетност (уеднаквяване)
- Стратегическо позициониране

## Ефективността не е стратегия

- Фокусът върху ключовите компетенции може да доведе компаниите до взаимно унищожителна конкуренция.
- Ефективността на операциите и стратегията са различни неща. Фирмите трябва да могат да ги разграничават.

## Ефективността не е стратегия

- Оперативната ефективност
- Бенчмаркинг
- Най-добри практики
- Продуктивност
- Ефикасност

## Ефективността не е стратегия

- Граница на продуктивността:
  - Максималната стойност, която компанията може да предостави на сътоветна цена с най-добрата технология, умения и мениджмънт.
  - Когато всички я увеличават, границата се отмества.
  - Всички трябва да се състезават с останалите, за да са близо до нея.

##  Ефективността не е стратегия

- Оперативната ефективност е важна, но не достатъчна за осигуряване на конкуретно предимство.
- Добрите практики се имитират лесно.
- Това създава конкурентна конвергентност – уеднаквяване на добрите практики.
- Изискванията за всички се увеличават.
- Конкурентите започват да си приличат – всички правят едно и също.
- Това е (чудесния) свят на перфектната конкуренция.

## Ефективността не е стратегия

- Оперативната ефективност е важна, но не достатъчна за осигуряване на конкуретно предимство.
- Конкуретното предимство трябва да бъде устойчиво – компанията трябва да прави нещо различно за продължително време.


##  Ефективността не е стратегия

- Оперативната ефективност означава сходни задачи да се изпълняват по-добре отколкото при конкурентите.
- Стратегическо позициониране означава да се правят различни неща или същите неща, но по различен начин.


## Конкурентна стратегия

- Стратегията е (в голяма степен) да изберете какво да не правите.
- Има ли НБУ стратегия?
- Имате ли вие стратегия?

## Конкурентна стратегия

- Ако конкуренцията е силна и правите това, което правят всички останали, правете го по различен начин.
- Създавайте добавена стойност по начин, който не може да се имитира лесно – т.е. изградете устойчиво предимство.


## Източници на добрата стратегия: позициониране

- Разнообразие
  - Определено множество от дейности (разнообразни клиенти малко потребности – смяна на масло; ниско-рискови инвестиции)
- Клиентски потребности
  - Задоволяване на повечето или всички потребности на една група клиенти (IKEA, Bessemer Trust, НБУ?)
- Достъпност
  - География, размер

## Стратегията е

- Създаване на уникална и стойностна позиция, включваща специфични действия – различни от тези на конкурентите.


## Компромиси

- Изработвайки стратегия, трябва (1) да се правят компромиси.
- Избираме една дейност вместо друга.
- Без компромиси всяка стратегия е лесна за имитация.

##  Комбиниране на дейности

- ..и (2) да се засилват съответствията при комбинирането на дейности.
- Стратегията е комбиниране на дейности в система.
- Комбинирането ограничава имитаторите, като изгражда уникална верига от дейности.
- Примери: low-cost авиолинии, IKEA, MIT, ... НБУ?
